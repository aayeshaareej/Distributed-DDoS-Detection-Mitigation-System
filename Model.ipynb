```python
import pandas as pd
from itertools import zip_longest

csv_files = [
    "DrDoS_DNS.csv",
    "DrDoS_LDAP.csv", 
    "DrDoS_MSSQL.csv",
    "DrDoS_NetBIOS.csv",
    "DrDoS_NTP.csv",
    "DrDoS_SNMP.csv",
    "DrDoS_SSDP.csv",
    "DrDoS_UDP.csv"
]

def side_by_side_comparison(file_list):
    print("=== SIDE-BY-SIDE COLUMN COMPARISON ===\n")
    
    # Collect headers
    headers_data = {}
    max_columns = 0
    
    for file in file_list:
        try:
            df = pd.read_csv(file, nrows=0)
            headers = df.columns.tolist()
            headers_data[file] = headers
            max_columns = max(max_columns, len(headers))
        except Exception as e:
            print(f"Error reading {file}: {e}")
            headers_data[file] = []
    
    # Create comparison table
    print(f"{'Column Index':<15}", end="")
    for file in file_list:
        print(f"{file:<25}", end="")
    print()
    
    print("-" * (15 + 25 * len(file_list)))
    
    for i in range(max_columns):
        print(f"{i:<15}", end="")
        for file in file_list:
            headers = headers_data[file]
            if i < len(headers):
                col_name = headers[i]
                print(f"{col_name:<25}", end="")
            else:
                print(f"{'':<25}", end="")
        print()
    
    # Check consistency
    print(f"\n{'='*50}")
    all_headers = [set(headers_data[f]) for f in file_list if headers_data[f]]
    
    if len(all_headers) > 1:
        all_identical = all(h == all_headers[0] for h in all_headers[1:])
        
        if all_identical:
            print("‚úÖ RESULT: All files have identical column structure")
        else:
            print("‚ùå RESULT: Files have different column structures")
            
            # Show unique columns per file
            common_cols = set.intersection(*all_headers)
            print(f"Common columns: {len(common_cols)}")
            
            for file in file_list:
                if headers_data[file]:
                    unique_cols = set(headers_data[file]) - common_cols
                    if unique_cols:
                        print(f"  {file} unique columns: {list(unique_cols)}")

# Run side-by-side comparison
side_by_side_comparison(csv_files)
```

```python
import pandas as pd
import os

# List of all CSV files
csv_files = [
    "DrDoS_DNS.csv",
    "DrDoS_LDAP.csv", 
    "DrDoS_MSSQL.csv",
    "DrDoS_NetBIOS.csv",
    "DrDoS_NTP.csv",
    "DrDoS_SNMP.csv",
    "DrDoS_SSDP.csv",
    "DrDoS_UDP.csv"
]

def process_and_compare_files(file_list):
    """
    Process all CSV files: clean headers, save them, and compare headers
    """
    headers_dict = {}
    processed_files = []
    
    print("üîÑ PROCESSING FILES AND CLEANING HEADERS")
    print("=" * 60)
    
    # Process each file
    for file in file_list:
        try:
            print(f"\nüìÅ Processing: {file}")
            
            # Read the CSV file
            df = pd.read_csv(file, low_memory=False)
            original_headers = df.columns.tolist()
            
            # Remove leading/trailing spaces from all column names
            df.columns = df.columns.str.strip()
            cleaned_headers = df.columns.tolist()
            
            # Save back to the same file (overwrite)
            df.to_csv(file, index=False)
            
            # Store headers for comparison
            headers_dict[file] = cleaned_headers
            processed_files.append(file)
            
            print(f"‚úÖ Changes saved back to {file}")
            print(f"üìä Original columns: {len(original_headers)}")
            print(f"üìä Cleaned columns: {len(cleaned_headers)}")
            
            # Show first 5 cleaned headers
            print(f"üîπ First 5 cleaned headers: {cleaned_headers[:5]}")
            
            # Count changes
            changes = sum(1 for orig, clean in zip(original_headers, cleaned_headers) if orig != clean)
            print(f"üîÑ Columns cleaned: {changes}")
            
        except Exception as e:
            print(f"‚ùå Error processing {file}: {e}")
            headers_dict[file] = None
    
    print("\n" + "=" * 60)
    print("üìã HEADERS COMPARISON ACROSS ALL FILES")
    print("=" * 60)
    
    # Compare headers across all files
    if headers_dict:
        compare_headers(headers_dict, processed_files)
    
    return headers_dict

def compare_headers(headers_dict, processed_files):
    """
    Compare headers across all processed files
    """
    # Get the first file as reference
    reference_file = processed_files[0]
    reference_headers = headers_dict[reference_file]
    
    print(f"\nReference file: {reference_file}")
    print(f"Number of columns: {len(reference_headers)}")
    
    # Check if all files have same headers
    all_same = True
    different_files = []
    
    for file in processed_files[1:]:
        current_headers = headers_dict[file]
        
        if current_headers == reference_headers:
            print(f"‚úÖ {file}: SAME headers as reference")
        else:
            print(f"‚ùå {file}: DIFFERENT headers from reference")
            all_same = False
            different_files.append(file)
            
            # Show differences
            ref_set = set(reference_headers)
            curr_set = set(current_headers)
            
            missing = ref_set - curr_set
            extra = curr_set - ref_set
            
            if missing:
                print(f"   Missing columns: {list(missing)}")
            if extra:
                print(f"   Extra columns: {list(extra)}")
    
    # Print summary
    print(f"\n" + "=" * 60)
    print("üìä COMPARISON SUMMARY")
    print("=" * 60)
    
    if all_same:
        print("üéâ ALL FILES HAVE IDENTICAL HEADERS!")
        print(f"   Total columns: {len(reference_headers)}")
        print(f"   Files processed: {len(processed_files)}")
    else:
        print(f"‚ö†Ô∏è  HEADERS ARE NOT IDENTICAL ACROSS ALL FILES")
        print(f"   Files with same headers: {len(processed_files) - len(different_files)}")
        print(f"   Files with different headers: {len(different_files)}")
        print(f"   Different files: {different_files}")
    
    # Show complete headers for reference file
    print(f"\nüìÑ COMPLETE HEADERS FOR {reference_file}:")
    print("-" * 40)
    for i, header in enumerate(reference_headers, 1):
        print(f"{i:3d}. {header}")

def display_all_headers(headers_dict):
    """
    Display all headers for each file in a formatted way
    """
    print(f"\n" + "=" * 60)
    print("üìÑ ALL HEADERS FOR EACH FILE")
    print("=" * 60)
    
    for file, headers in headers_dict.items():
        if headers is not None:
            print(f"\nüìÅ {file} ({len(headers)} columns):")
            print("-" * 40)
            # Show first 10 headers
            for i, header in enumerate(headers[:10], 1):
                print(f"  {i:2d}. {header}")
            if len(headers) > 10:
                print(f"  ... and {len(headers) - 10} more columns")
        else:
            print(f"\n‚ùå {file}: Could not read headers")

# Main execution
if __name__ == "__main__":
    print("üöÄ STARTING CSV HEADER CLEANING AND COMPARISON")
    print("=" * 60)
    
    # Process all files and get cleaned headers
    headers_dict = process_and_compare_files(csv_files)
    
    # Display all headers
    display_all_headers(headers_dict)
    
    # Final summary
    print(f"\n" + "=" * 60)
    print("üéØ FINAL SUMMARY")
    print("=" * 60)
    
    successful_files = [f for f in csv_files if headers_dict.get(f) is not None]
    failed_files = [f for f in csv_files if headers_dict.get(f) is None]
    
    print(f"‚úÖ Successfully processed: {len(successful_files)} files")
    print(f"‚ùå Failed to process: {len(failed_files)} files")
    
    if successful_files:
        print(f"üìÅ Successful files: {successful_files}")
    if failed_files:
        print(f"üìÅ Failed files: {failed_files}")
```

```python
import pandas as pd

# ----------------------------
# SETTINGS
# ----------------------------
attack_samples_total = 200000   # total attack rows you want
attack_files = [
    "DrDoS_DNS.csv",
    "DrDoS_LDAP.csv",
    "DrDoS_MSSQL.csv",
    "DrDoS_NetBIOS.csv",
    "DrDoS_NTP.csv",
    "DrDoS_SNMP.csv",
    "DrDoS_SSDP.csv",
    "DrDoS_UDP.csv"
]

# How many samples per attack class?
attack_samples_per_file = attack_samples_total // len(attack_files)

print("Sampling", attack_samples_per_file, "from each attack file")

# ----------------------------
# LOAD & SAMPLE ATTACK ROWS
# ----------------------------
attack_list = []
benign_list = []

for f in attack_files:
    print("Processing:", f)

    df = pd.read_csv(f, low_memory=False)

    # Attack = Label != BENIGN
    attack_df = df[df[" Label"] != "BENIGN"]

    if len(attack_df) < attack_samples_per_file:
        print(f"‚ö† Warning: {f} has fewer attack rows than needed ({len(attack_df)}). Sampling all available.")
        sampled_attack = attack_df
    else:
        sampled_attack = attack_df.sample(n=attack_samples_per_file, random_state=42)

    attack_list.append(sampled_attack)

    # Also store benign rows from this file
    benign_df = df[df[" Label"] == "BENIGN"]
    if len(benign_df) > 0:
        benign_list.append(benign_df)

# ----------------------------
# COMBINE ALL BENIGN ROWS
# ----------------------------
all_benign_df = pd.concat(benign_list, ignore_index=True)

print("\nTotal benign rows found:", len(all_benign_df))

# ----------------------------
# CREATE FINAL DATASET
# ----------------------------
final_attack_df = pd.concat(attack_list, ignore_index=True)

final_df = pd.concat([final_attack_df, all_benign_df], ignore_index=True)

# Shuffle dataset
final_df = final_df.sample(frac=1, random_state=42).reset_index(drop=True)

# ----------------------------
# SAVE
# ----------------------------
final_df.to_csv("DDoS_Enhanced_Dataset.csv", index=False)

print("\n--------------------------------------------")
print("DATASET CREATED SUCCESSFULLY!")
print("Attack rows:", len(final_attack_df))
print("Benign rows:", len(all_benign_df))
print("Total rows:", len(final_df))
print("Saved as: DDoS_Enhanced_Dataset.csv")
print("--------------------------------------------")
```

```python
import pandas as pd
import time

# Try to import cuDF for GPU acceleration
try:
    import cudf
    CUDA_AVAILABLE = True
    print("üéâ USING GPU ACCELERATION (CUDA)")
except ImportError:
    CUDA_AVAILABLE = False
    print("üíª USING CPU PROCESSING (pandas)")

# List of CSV files
csv_files = [
    "DrDoS_DNS.csv",
    "DrDoS_LDAP.csv", 
    "DrDoS_MSSQL.csv",
    "DrDoS_NetBIOS.csv",
    "DrDoS_NTP.csv",
    "DrDoS_SNMP.csv",
    "DrDoS_SSDP.csv",
    "DrDoS_UDP.csv",
    "Syn.csv",
    "TFTP.csv",
    "UDPLag.csv",
    "LDAP.csv",
    "MSSQL.csv",
    "NetBIOS.csv",
    "Portmap.csv",
    "UDP.csv"
]

def read_csv_gpu(file_path):
    """Read CSV using GPU if available"""
    if CUDA_AVAILABLE:
        return cudf.read_csv(file_path, low_memory=False)
    else:
        return pd.read_csv(file_path, low_memory=False)

# Collect all BENIGN traffic with GPU acceleration
benign_data = []
total_benign = 0
start_time = time.time()

print(f"üöÄ PROCESSING {len(csv_files)} FILES WITH {'GPU' if CUDA_AVAILABLE else 'CPU'}")
print("=" * 50)

for file in csv_files:
    try:
        file_start = time.time()
        
        # Read file with GPU/CPU
        df = read_csv_gpu(file)
        
        # Filter BENIGN traffic (case insensitive)
        # Note: Column name has space ' Label' based on your previous code
        benign_mask = df[' Label'].str.contains('BENIGN', case=False, na=False)
        benign_rows = df[benign_mask]
        
        # Convert to pandas if using GPU (for concatenation)
        if CUDA_AVAILABLE:
            benign_rows = benign_rows.to_pandas()
        
        processing_time = time.time() - file_start
        print(f"üìÅ {file}: {len(benign_rows)} BENIGN rows [{processing_time:.2f}s]")
        
        benign_data.append(benign_rows)
        total_benign += len(benign_rows)
        
    except Exception as e:
        print(f"‚ùå Error processing {file}: {e}")

# Combine all BENIGN data
if benign_data:
    all_benign = pd.concat(benign_data, ignore_index=True)
    
    # Save to new file
    all_benign.to_csv('ALL_BENIGN_TRAFFIC.csv', index=False)
    
    total_time = time.time() - start_time
    print(f"\n‚úÖ Saved {len(all_benign):,} BENIGN rows to 'ALL_BENIGN_TRAFFIC.csv'")
    print(f"‚è±Ô∏è  Total processing time: {total_time:.2f} seconds")
    print(f"üìä Average time per file: {total_time/len(csv_files):.2f} seconds")
    print(f"üéØ Processing method: {'GPU (CUDA)' if CUDA_AVAILABLE else 'CPU (pandas)'}")
else:
    print("‚ùå No BENIGN data found in any files")
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix, classification_report
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.utils.class_weight import compute_class_weight
import warnings
warnings.filterwarnings('ignore')

# Check if CUDA is available
try:
    import cupy as cp
    from cuml.ensemble import RandomForestClassifier as cuRF
    CUDA_AVAILABLE = True
    print("‚úÖ CUDA is available - Using GPU acceleration")
except ImportError:
    CUDA_AVAILABLE = False
    print("‚ö†Ô∏è CUDA not available - Using CPU")

class DDoSDetector:
    def __init__(self):
        self.label_encoder = None
        self.scaler = StandardScaler()
        self.model = None
        self.CUDA_AVAILABLE = CUDA_AVAILABLE
        
    def load_and_preprocess_data(self, file_path):
        """Load and preprocess the dataset"""
        print("üìä Loading dataset...")
        df = pd.read_csv(file_path)
        
        # Print dataset info
        print(f"Dataset shape: {df.shape}")
        print(f"BENIGN rows: {len(df[df[' Label'] == 'BENIGN'])}")
        print(f"Attack rows: {len(df[df[' Label'] != 'BENIGN'])}")
        print(f"Total rows: {len(df)}")
        
        # Remove unnecessary columns that can cause static learning
        columns_to_drop = [
            'Unnamed: 0', 'Flow ID', ' Source IP', ' Destination IP', 
            ' Source Port', ' Destination Port', ' Timestamp', 'Flow ID'
        ]
        
        # Only drop columns that exist in the dataframe
        columns_to_drop = [col for col in columns_to_drop if col in df.columns]
        df_clean = df.drop(columns=columns_to_drop)
        
        print(f"Features after cleaning: {len(df_clean.columns)}")
        
        # Handle missing values and infinities
        df_clean = df_clean.replace([np.inf, -np.inf], np.nan)
        
        # Fill missing values with median for numerical columns
        numerical_cols = df_clean.select_dtypes(include=[np.number]).columns
        for col in numerical_cols:
            if df_clean[col].isna().any():
                df_clean[col].fillna(df_clean[col].median(), inplace=True)
        
        # Encode the target variable
        self.label_encoder = LabelEncoder()
        df_clean['Label_encoded'] = self.label_encoder.fit_transform(df_clean[' Label'])
        
        # Print class distribution
        print("\nüéØ Class Distribution:")
        print("=" * 40)
        for class_name in self.label_encoder.classes_:
            count = len(df_clean[df_clean[' Label'] == class_name])
            percentage = (count / len(df_clean)) * 100
            print(f"{class_name:<15}: {count:>6} ({percentage:.2f}%)")
        
        return df_clean
    
    def prepare_features(self, df):
        """Prepare features and target variable"""
        # Drop the original label column and use encoded one
        X = df.drop(columns=[' Label', 'Label_encoded'])
        y = df['Label_encoded']
        
        # Ensure all features are numeric
        for col in X.columns:
            if X[col].dtype == 'object':
                try:
                    X[col] = pd.to_numeric(X[col], errors='coerce')
                except:
                    X[col] = LabelEncoder().fit_transform(X[col])
        
        # Final check for any remaining NaN values
        X = X.fillna(0)
        
        print(f"\nüìà Feature matrix shape: {X.shape}")
        print(f"üéØ Target variable shape: {y.shape}")
        
        return X, y
    
    def calculate_detailed_metrics(self, y_true, y_pred):
        """Calculate comprehensive metrics"""
        # Overall metrics
        precision = precision_score(y_true, y_pred, average='weighted', zero_division=0)
        recall = recall_score(y_true, y_pred, average='weighted', zero_division=0)
        f1 = f1_score(y_true, y_pred, average='weighted', zero_division=0)
        
        # Calculate False Positive Rate
        cm = confusion_matrix(y_true, y_pred)
        fp = cm.sum(axis=0) - np.diag(cm)
        tn = cm.sum() - (cm.sum(axis=1) + fp)
        fpr = fp / (fp + tn + 1e-8)  # Add small value to avoid division by zero
        fpr_mean = fpr.mean()
        
        return precision, recall, f1, fpr_mean, cm
    
    def print_metrics(self, y_true, y_pred, cm):
        """Print comprehensive metrics"""
        precision, recall, f1, fpr_mean, _ = self.calculate_detailed_metrics(y_true, y_pred)
        
        print("\n" + "="*70)
        print("üéØ ACCURACY METRICS")
        print("="*70)
        print(f"üìç Precision (Weighted):  {precision:.4f}")
        print(f"üìä Recall (Weighted):     {recall:.4f}")
        print(f"‚ö° F1-Score (Weighted):   {f1:.4f}")
        print(f"üö´ False Positive Rate:   {fpr_mean:.4f}")
        
        # Detailed class-wise metrics
        print("\n" + "="*70)
        print("üìã DETAILED CLASS-WISE METRICS")
        print("="*70)
        
        class_names = self.label_encoder.classes_
        
        for i, class_name in enumerate(class_names):
            # Calculate class-specific metrics
            class_mask = (y_true == i)
            if np.sum(class_mask) > 0:  # Only calculate if class exists in test set
                class_precision = precision_score(y_true, y_pred, average=None, zero_division=0)[i]
                class_recall = recall_score(y_true, y_pred, average=None, zero_division=0)[i]
                class_f1 = f1_score(y_true, y_pred, average=None, zero_division=0)[i]
                
                # Calculate class-specific FPR
                tp = cm[i, i]
                fp = cm[:, i].sum() - tp
                tn = cm.sum() - (cm[i, :].sum() + cm[:, i].sum() - tp)
                class_fpr = fp / (fp + tn + 1e-8)
                
                print(f"\nüîπ {class_name}:")
                print(f"   Precision:  {class_precision:.4f}")
                print(f"   Recall:     {class_recall:.4f}")
                print(f"   F1-Score:   {class_f1:.4f}")
                print(f"   FPR:        {class_fpr:.4f}")
        
        # Print confusion matrix summary
        print("\n" + "="*70)
        print("üìä CONFUSION MATRIX SUMMARY")
        print("="*70)
        print(f"True Positives:  {np.diag(cm).sum()}")
        print(f"False Positives: {(cm.sum(axis=0) - np.diag(cm)).sum()}")
        print(f"False Negatives: {(cm.sum(axis=1) - np.diag(cm)).sum()}")
        print(f"Total Samples:   {cm.sum()}")
        
        return precision, recall, f1, fpr_mean
    
    def train_model(self, X_train, X_test, y_train, y_test):
        """Train Random Forest model with class weighting"""
        print("\nüöÄ Training Random Forest Model...")
        print(f"Training set: {X_train.shape[0]} samples")
        print(f"Test set: {X_test.shape[0]} samples")
        
        if self.CUDA_AVAILABLE:
            print("üéØ Using GPU-accelerated Random Forest (cuML)")
            
            # Convert to GPU arrays
            X_train_gpu = cp.asarray(X_train.astype(np.float32))
            X_test_gpu = cp.asarray(X_test.astype(np.float32))
            y_train_gpu = cp.asarray(y_train.astype(np.int32))
            y_test_gpu = cp.asarray(y_test.astype(np.int32))
            
            # Initialize and train GPU Random Forest with class weighting
            self.model = cuRF(
                n_estimators=150,
                max_depth=25,
                random_state=42,
                class_weight='balanced',  # Handle class imbalance
                n_streams=2
            )
            
            self.model.fit(X_train_gpu, y_train_gpu)
            
            # Predictions
            y_pred = self.model.predict(X_test_gpu)
            y_pred = cp.asnumpy(y_pred)
            y_test_cpu = cp.asnumpy(y_test_gpu)
            
        else:
            print("üéØ Using CPU Random Forest (scikit-learn)")
            
            # Calculate class weights explicitly for imbalance handling
            class_weights = compute_class_weight(
                'balanced',
                classes=np.unique(y_train),
                y=y_train
            )
            weight_dict = dict(enumerate(class_weights))
            
            print("üìä Class weights for imbalance handling:")
            for class_id, weight in weight_dict.items():
                class_name = self.label_encoder.inverse_transform([class_id])[0]
                print(f"   {class_name}: {weight:.4f}")
            
            # Initialize and train CPU Random Forest
            self.model = RandomForestClassifier(
                n_estimators=150,
                max_depth=25,
                random_state=42,
                class_weight=weight_dict,  # Explicit class weights
                n_jobs=-1,  # Use all available cores
                bootstrap=True,
                min_samples_split=5,
                min_samples_leaf=2
            )
            
            self.model.fit(X_train, y_train)
            
            # Predictions
            y_pred = self.model.predict(X_test)
            y_test_cpu = y_test
        
        # Calculate and print metrics
        cm = confusion_matrix(y_test_cpu, y_pred)
        precision, recall, f1, fpr = self.print_metrics(y_test_cpu, y_pred, cm)
        
        # Feature importance (for CPU version)
        if not self.CUDA_AVAILABLE and hasattr(self.model, 'feature_importances_'):
            self.print_feature_importance(X_train.columns)
        
        return precision, recall, f1, fpr
    
    def print_feature_importance(self, feature_names):
        """Print top 15 most important features"""
        print("\n" + "="*70)
        print("üîç TOP 15 MOST IMPORTANT FEATURES")
        print("="*70)
        
        feature_importance = pd.DataFrame({
            'Feature': feature_names,
            'Importance': self.model.feature_importances_
        }).sort_values('Importance', ascending=False)
        
        print(feature_importance.head(15).to_string(index=False))
    
    def run_complete_pipeline(self, file_path):
        """Run the complete DDoS detection pipeline"""
        try:
            # Step 1: Load and preprocess data
            df = self.load_and_preprocess_data(file_path)
            
            # Step 2: Prepare features
            X, y = self.prepare_features(df)
            
            # Step 3: Split data (70% train, 30% test) with stratification
            X_train, X_test, y_train, y_test = train_test_split(
                X, y, 
                test_size=0.3, 
                random_state=42, 
                stratify=y  # Maintain class distribution
            )
            
            # Step 4: Scale features
            print("\nüìê Scaling features...")
            X_train_scaled = self.scaler.fit_transform(X_train)
            X_test_scaled = self.scaler.transform(X_test)
            
            # Convert back to DataFrame to maintain column names
            X_train_df = pd.DataFrame(X_train_scaled, columns=X.columns)
            X_test_df = pd.DataFrame(X_test_scaled, columns=X.columns)
            
            # Step 5: Train model and get metrics
            precision, recall, f1, fpr = self.train_model(
                X_train_df, X_test_df, y_train, y_test
            )
            
            # Final summary
            print("\n" + "="*70)
            print("‚úÖ MODEL TRAINING COMPLETED SUCCESSFULLY!")
            print("="*70)
            print(f"üéØ Final Weighted Metrics:")
            print(f"   Precision:  {precision:.4f}")
            print(f"   Recall:     {recall:.4f}")
            print(f"   F1-Score:   {f1:.4f}")
            print(f"   FPR:        {fpr:.4f}")
            
            return self.model, precision, recall, f1, fpr
            
        except FileNotFoundError:
            print(f"‚ùå Error: File '{file_path}' not found.")
            return None
        except Exception as e:
            print(f"‚ùå Error during execution: {str(e)}")
            import traceback
            traceback.print_exc()
            return None

# Main execution
if __name__ == "__main__":
    # Initialize detector
    detector = DDoSDetector()
    
    # Update this path to your CSV file
    file_path = "DDoS_Enhanced_Dataset.csv"  # CHANGE THIS TO YOUR ACTUAL FILE PATH
    
    # Run complete pipeline
    results = detector.run_complete_pipeline(file_path)
    
    if results:
        print("\nüéâ DDoS Detection Model Ready!")
    else:
        print("\nüí• Pipeline execution failed. Please check the file path and data format.")
```

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")

# ----------------------------
# 0Ô∏è‚É£ CUDA check
# ----------------------------
cuda_available = False
try:
    from cuml.ensemble import RandomForestClassifier as cuRF
    cuda_available = True
    print("‚úî Using CUDA GPU with cuML RandomForest")
except:
    from sklearn.ensemble import RandomForestClassifier as skRF
    cuda_available = False
    print("‚úî cuML not available ‚Üí Using sklearn RandomForest (CPU)")

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix

# ----------------------------
# 1Ô∏è‚É£ Load dataset
# ----------------------------
df = pd.read_csv("DDoS_Enhanced_Dataset.csv", low_memory=False)
print("Dataset Loaded:", df.shape)

# ----------------------------
# 2Ô∏è‚É£ Drop static features
# ----------------------------
drop_cols = [
    'Unnamed: 0',
    'Flow ID',
    ' Source IP',
    ' Destination IP',
    ' Source Port',
    ' Destination Port',
    ' Timestamp'
]
df = df.drop(columns=drop_cols, errors="ignore")

# ----------------------------
# 3Ô∏è‚É£ Label encoding
# ----------------------------
le = LabelEncoder()
df["Label"] = le.fit_transform(df[" Label"].astype(str))

y = df["Label"]
X = df.drop(columns=["Label", " Label"], errors="ignore")

# ----------------------------
# 4Ô∏è‚É£ Clean features
# ----------------------------
# Convert to numeric, replace NaN and infinities, clip extreme values
X = X.apply(pd.to_numeric, errors='coerce')  # ensure numeric
X = X.replace([np.inf, -np.inf], np.nan)    # replace inf
X = X.fillna(0)                             # replace NaN
X = X.clip(-1e10, 1e10)                     # clip extreme values

# ----------------------------
# 5Ô∏è‚É£ Train/test split (70/30)
# ----------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.30, random_state=42, stratify=y
)

# Safety: clean again after split
X_train = X_train.replace([np.inf, -np.inf], np.nan).fillna(0).clip(-1e10, 1e10)
X_test  = X_test.replace([np.inf, -np.inf], np.nan).fillna(0).clip(-1e10, 1e10)

print("Train:", X_train.shape, " Test:", X_test.shape)

# ----------------------------
# 6Ô∏è‚É£ Weighted Random Forest
# ----------------------------
if cuda_available:
    # cuML: use sample_weight to handle class imbalance
    classes, counts = np.unique(y_train.to_numpy(), return_counts=True)
    total = len(y_train)
    class_weights = {c: total/count for c, count in zip(classes, counts)}
    sample_weights = np.array([class_weights[label] for label in y_train])
    
    model = cuRF(
        n_estimators=200,
        max_depth=20,
        n_streams=8
    )
    
    print("Training weighted RandomForest on GPU...")
    model.fit(X_train, y_train, sample_weight=sample_weights)

else:
    # sklearn: use class_weight='balanced'
    model = skRF(
        n_estimators=200,
        max_depth=20,
        n_jobs=-1,
        class_weight="balanced"
    )
    
    print("Training weighted RandomForest on CPU...")
    model.fit(X_train, y_train)

print("Training finished!")

# ----------------------------
# 7Ô∏è‚É£ Predict
# ----------------------------
y_pred = model.predict(X_test)
y_pred = np.array(y_pred).astype(int)

# ----------------------------
# 8Ô∏è‚É£ Metrics
# ----------------------------
print("\n\n===== CLASSIFICATION REPORT =====")
print(classification_report(y_test, y_pred, target_names=le.classes_))

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:\n", cm)

# ----------------------------
# 9Ô∏è‚É£ False Positive Rate (per class)
# ----------------------------
FPR = {}
for i, cls in enumerate(le.classes_):
    FP = cm[:, i].sum() - cm[i, i]
    TN = cm.sum() - (cm[i, :].sum() + cm[:, i].sum() - cm[i, i])
    FPR[cls] = FP / (FP + TN)

print("\n===== FALSE POSITIVE RATE (per class) =====")
for cls, rate in FPR.items():
    print(f"{cls}: {rate:.6f}")
```

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")

# ----------------------------
# 0Ô∏è‚É£ CUDA check
# ----------------------------
cuda_available = False
try:
    from cuml.ensemble import RandomForestClassifier as cuRF
    from cuml.multioutput import MultiOutputClassifier as cuMultiOutput
    cuda_available = True
    print("‚úî Using CUDA GPU with cuML RandomForest")
except:
    from sklearn.ensemble import RandomForestClassifier as skRF
    from sklearn.multioutput import MultiOutputClassifier
    cuda_available = False
    print("‚úî cuML not available ‚Üí Using sklearn RandomForest (CPU)")

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.metrics import classification_report, confusion_matrix

# ----------------------------
# 1Ô∏è‚É£ Load dataset
# ----------------------------
df = pd.read_csv("DDoS_Enhanced_Dataset.csv", low_memory=False)
print("Dataset Loaded:", df.shape)

# ----------------------------
# 2Ô∏è‚É£ Drop static features
# ----------------------------
drop_cols = [
    'Unnamed: 0',
    'Flow ID',
    ' Source IP',
    ' Destination IP',
    ' Source Port',
    ' Destination Port',
    ' Timestamp'
]
df = df.drop(columns=drop_cols, errors="ignore")

# ----------------------------
# 3Ô∏è‚É£ Multi-label encoding
# ----------------------------
# Assume labels are comma-separated, e.g., "DoS,Botnet"
mlb = MultiLabelBinarizer()
y = mlb.fit_transform(df[" Label"].astype(str).str.split(","))

# Features
X = df.drop(columns=[" Label"], errors="ignore")

# ----------------------------
# 4Ô∏è‚É£ Clean features
# ----------------------------
X = X.apply(pd.to_numeric, errors='coerce')      # ensure numeric
X = X.replace([np.inf, -np.inf], np.nan)        # replace inf
X = X.fillna(0)                                 # replace NaN
X = X.clip(-1e10, 1e10)                         # clip extreme values

# ----------------------------
# 5Ô∏è‚É£ Train/test split
# ----------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.30, random_state=42, stratify=None  # stratify cannot be used for multi-label
)

# Safety: clean again after split
X_train = X_train.replace([np.inf, -np.inf], np.nan).fillna(0).clip(-1e10, 1e10)
X_test  = X_test.replace([np.inf, -np.inf], np.nan).fillna(0).clip(-1e10, 1e10)

print("Train:", X_train.shape, " Test:", X_test.shape)

# ----------------------------
# 6Ô∏è‚É£ Weighted Random Forest
# ----------------------------
if cuda_available:
    # cuML multi-output RF
    classes_per_label = y_train.sum(axis=0)
    # compute sample weights per label
    sample_weights = np.ones(len(y_train))
    # simple weighting: sum over labels
    for i, count in enumerate(classes_per_label):
        sample_weights += (y_train[:, i] == 1) * (len(y_train) / count - 1)
    
    base_model = cuRF(
        n_estimators=200,
        max_depth=20,
        n_streams=8
    )
    model = cuMultiOutput(base_model)
    
    print("Training weighted Multi-Label RandomForest on GPU...")
    model.fit(X_train.values, y_train, sample_weight=sample_weights)

else:
    # sklearn multi-output RF
    base_model = skRF(
        n_estimators=200,
        max_depth=20,
        n_jobs=-1,
        class_weight="balanced"
    )
    model = MultiOutputClassifier(base_model, n_jobs=-1)
    
    print("Training weighted Multi-Label RandomForest on CPU...")
    model.fit(X_train, y_train)

print("Training finished!")

# ----------------------------
# 7Ô∏è‚É£ Predict
# ----------------------------
y_pred = model.predict(X_test)
y_pred = np.array(y_pred).astype(int)

# ----------------------------
# 8Ô∏è‚É£ Metrics
# ----------------------------
print("\n\n===== CLASSIFICATION REPORT =====")
print(classification_report(y_test, y_pred, target_names=mlb.classes_))

# Confusion matrix per label
print("\n===== CONFUSION MATRICES (per label) =====")
for i, cls in enumerate(mlb.classes_):
    cm = confusion_matrix(y_test[:, i], y_pred[:, i])
    print(f"\nClass: {cls}\n{cm}")

# ----------------------------
# 9Ô∏è‚É£ False Positive Rate (per label)
# ----------------------------
FPR = {}
for i, cls in enumerate(mlb.classes_):
    cm = confusion_matrix(y_test[:, i], y_pred[:, i])
    FP = cm[0, 1] if cm.shape == (2,2) else 0
    TN = cm[0, 0] if cm.shape == (2,2) else 0
    FPR[cls] = FP / (FP + TN) if (FP + TN) > 0 else 0

print("\n===== FALSE POSITIVE RATE (per label) =====")
for cls, rate in FPR.items():
    print(f"{cls}: {rate:.6f}")
```

```python
import pandas as pd
import numpy as np
import warnings
import matplotlib.pyplot as plt
import seaborn as sns
import joblib

warnings.filterwarnings("ignore")

# ----------------------------
# 0Ô∏è‚É£ Imports
# ----------------------------
from sklearn.ensemble import RandomForestClassifier
from sklearn.multioutput import MultiOutputClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_recall_fscore_support, confusion_matrix, f1_score
from sklearn.utils.class_weight import compute_sample_weight
from imblearn.over_sampling import SMOTE

# ----------------------------
# 1Ô∏è‚É£ Load dataset
# ----------------------------
df = pd.read_csv("DDoS_Enhanced_Dataset.csv", low_memory=False)
print("Dataset Loaded:", df.shape)

# ----------------------------
# 2Ô∏è‚É£ Drop static columns
# ----------------------------
drop_cols = [
    'Unnamed: 0', 'Flow ID', ' Source IP', ' Destination IP',
    ' Source Port', ' Destination Port', ' Timestamp'
]
df = df.drop(columns=drop_cols, errors="ignore")

# ----------------------------
# 3Ô∏è‚É£ Multi-label target
# ----------------------------
multi_labels = [
    'DrDoS_SSDP', 'BENIGN', 'DrDoS_UDP', 'DrDoS_MSSQL',
    'DrDoS_NTP', 'DrDoS_DNS', 'DrDoS_LDAP', 'DrDoS_SNMP'
]

for label in multi_labels:
    df[label] = (df[" Label"].str.contains(label)).astype(int)

y = df[multi_labels].values
X = df.drop(columns=[" Label"] + multi_labels, errors="ignore")

# ----------------------------
# 4Ô∏è‚É£ Clean features
# ----------------------------
X = X.apply(pd.to_numeric, errors='coerce').fillna(0)
X = X.clip(-1e10, 1e10)

# ----------------------------
# 5Ô∏è‚É£ Train/test split
# ----------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=None
)

# ----------------------------
# 6Ô∏è‚É£ Multi-label SMOTE (oversample minority labels)
# ----------------------------
# Create a binary indicator: 1 if any rare label exists
rare_indicator = (y_train.sum(axis=1) > 0).astype(int)
smote = SMOTE(sampling_strategy=1.0, random_state=42)
X_train_res, rare_indicator_res = smote.fit_resample(X_train, rare_indicator)

# Replicate corresponding y labels for oversampled rows
# Get indices where rare_indicator == 1
X_rare = X_train[rare_indicator==1]
y_rare = y_train[rare_indicator==1]

X_train_res = pd.concat([pd.DataFrame(X_train), pd.DataFrame(X_rare)], ignore_index=True)
y_train_res = np.vstack([y_train, y_rare])

X_train = X_train_res
y_train = y_train_res

print("After SMOTE, X_train:", X_train.shape, "y_train:", y_train.shape)

# ----------------------------
# 7Ô∏è‚É£ Sample weights for imbalance
# ----------------------------
sample_weights = np.ones(len(y_train))
for i in range(len(multi_labels)):
    weights = compute_sample_weight(class_weight='balanced', y=y_train[:, i])
    sample_weights += weights

# ----------------------------
# 8Ô∏è‚É£ Train Multi-label Random Forest
# ----------------------------
base_model = RandomForestClassifier(
    n_estimators=200, max_depth=20, n_jobs=-1, class_weight='balanced'
)
model = MultiOutputClassifier(base_model, n_jobs=-1)
print("Training Multi-Label RandomForest...")
model.fit(X_train, y_train)

print("Training finished!")

# ----------------------------
# 9Ô∏è‚É£ Predict probabilities
# ----------------------------
y_pred_proba = np.array([est.predict_proba(X_test)[:,1] for est in model.estimators_]).T

# ----------------------------
# üîü Per-label threshold optimization
# ----------------------------
optimal_thresholds = []
for i in range(len(multi_labels)):
    best_thresh = 0.5
    best_f1 = 0
    for t in np.arange(0.1, 0.9, 0.05):
        y_pred_label = (y_pred_proba[:,i] >= t).astype(int)
        f1 = f1_score(y_test[:,i], y_pred_label)
        if f1 > best_f1:
            best_f1 = f1
            best_thresh = t
    optimal_thresholds.append(best_thresh)

y_pred = np.zeros_like(y_pred_proba, dtype=int)
for i, t in enumerate(optimal_thresholds):
    y_pred[:, i] = (y_pred_proba[:, i] >= t).astype(int)

print("Optimal thresholds per label:", dict(zip(multi_labels, optimal_thresholds)))

# ----------------------------
# 1Ô∏è‚É£1Ô∏è‚É£ Metrics per label
# ----------------------------
precision, recall, f1, _ = precision_recall_fscore_support(y_test, y_pred, average=None)
accuracy = (y_test == y_pred).mean(axis=0)

FPR = []
for i in range(len(multi_labels)):
    cm = confusion_matrix(y_test[:, i], y_pred[:, i])
    FP = cm[0,1] if cm.shape==(2,2) else 0
    TN = cm[0,0] if cm.shape==(2,2) else 0
    FPR.append(FP / (FP + TN) if (FP + TN) > 0 else 0)

metrics_df = pd.DataFrame({
    "Label": multi_labels,
    "Precision": precision,
    "Recall": recall,
    "F1": f1,
    "Accuracy": accuracy,
    "FPR": FPR
})
print(metrics_df)

# ----------------------------
# 1Ô∏è‚É£2Ô∏è‚É£ Bar chart of metrics
# ----------------------------
plt.figure(figsize=(12,6))
metrics_melt = metrics_df.melt(id_vars="Label", var_name="Metric", value_name="Value")
sns.barplot(x="Label", y="Value", hue="Metric", data=metrics_melt)
plt.title("Multi-Label Random Forest Performance Metrics (After SMOTE + Threshold Optimization)")
plt.xticks(rotation=45)
plt.ylim(0,1.0)
plt.legend(loc="upper right")
plt.tight_layout()
plt.show()

# ----------------------------
# 1Ô∏è‚É£3Ô∏è‚É£ Confusion matrix heatmaps per label
# ----------------------------
for i, label in enumerate(multi_labels):
    cm = confusion_matrix(y_test[:, i], y_pred[:, i])
    plt.figure(figsize=(4,3))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False)
    plt.title(f"Confusion Matrix: {label}")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()

# ----------------------------
# 1Ô∏è‚É£4Ô∏è‚É£ Overall TP/FP/FN/TN pie chart
# ----------------------------
TP = FP = FN = TN = 0
for i in range(len(multi_labels)):
    cm = confusion_matrix(y_test[:, i], y_pred[:, i])
    if cm.shape == (2,2):
        TN += cm[0,0]
        FP += cm[0,1]
        FN += cm[1,0]
        TP += cm[1,1]

plt.figure(figsize=(6,6))
sizes = [TP, FP, FN, TN]
labels_pie = ["True Positive", "False Positive", "False Negative", "True Negative"]
colors = ["#4CAF50", "#FF5722", "#FFC107", "#2196F3"]
plt.pie(sizes, labels=labels_pie, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title("Overall Prediction Distribution Across All Labels (SMOTE + Threshold Optimized)")
plt.show()

# ----------------------------
# 1Ô∏è‚É£5Ô∏è‚É£ Save the model
# ----------------------------
joblib.dump(model, "multi_label_rf_model_smote_opt.pkl")
print("Model saved as 'multi_label_rf_model_smote_opt.pkl'")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import (classification_report, confusion_matrix, accuracy_score, 
                           precision_score, recall_score, f1_score, precision_recall_fscore_support)
import warnings
warnings.filterwarnings('ignore')

# Set style for better plots
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

# Load your balanced dataset
df = pd.read_csv('balanced_dataset.csv')

# Display dataset info
print("Dataset shape:", df.shape)
print("\nLabel distribution:")
print(df[' Label'].value_counts())

# Identify columns to exclude
exclude_columns = [
    'Unnamed: 0',      # Index column
    'Flow ID',         # Flow identifier
    ' Source IP',      # IP address
    ' Destination IP', # IP address
    ' Timestamp',      # Timestamp
    ' Label'           # This is our target variable
]

# Features to use (all remaining columns except the excluded ones)
feature_columns = [col for col in df.columns if col not in exclude_columns]

print(f"\nUsing {len(feature_columns)} features:")

# Prepare features and target
X = df[feature_columns]
y = df[' Label']

# Data Cleaning Steps
print("\n=== DATA CLEANING ===")

# 1. Convert non-numeric columns to numeric
non_numeric_cols = X.select_dtypes(include=['object']).columns.tolist()
if non_numeric_cols:
    print(f"Converting non-numeric columns: {non_numeric_cols}")
    for col in non_numeric_cols:
        X[col] = pd.to_numeric(X[col], errors='coerce')

# 2. Check for infinite values and replace them
print("Checking for infinite values...")
inf_mask = np.isinf(X.select_dtypes(include=[np.number]))
if inf_mask.any().any():
    inf_columns = inf_mask.any()
    inf_cols_list = inf_columns[inf_columns].index.tolist()
    print(f"Found infinite values in columns: {inf_cols_list}")
    
    # Replace inf with NaN
    X = X.replace([np.inf, -np.inf], np.nan)

# 3. Handle extreme values by clipping
print("Handling extreme values...")
for col in X.select_dtypes(include=[np.number]).columns:
    Q1 = X[col].quantile(0.25)
    Q3 = X[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 3 * IQR
    upper_bound = Q3 + 3 * IQR
    
    # Clip extreme outliers
    if not np.isnan(lower_bound) and not np.isnan(upper_bound):
        outliers_before = ((X[col] < lower_bound) | (X[col] > upper_bound)).sum()
        if outliers_before > 0:
            print(f"  {col}: clipping {outliers_before} outliers")
            X[col] = np.clip(X[col], lower_bound, upper_bound)

# 4. Fill NaN values with median
print("Filling NaN values...")
nan_before = X.isnull().sum().sum()
print(f"NaN values before filling: {nan_before}")

for col in X.columns:
    if X[col].isnull().sum() > 0:
        X[col].fillna(X[col].median(), inplace=True)

print(f"NaN values after filling: {X.isnull().sum().sum()}")

# 5. Scale large values to prevent float32 overflow
print("Scaling large values...")
for col in X.select_dtypes(include=[np.number]).columns:
    col_max = abs(X[col]).max()
    if col_max > 1e6:
        print(f"  Scaling {col} (max absolute value: {col_max:.2e})")
        X[col] = X[col] / col_max

# Convert to float32
X = X.astype(np.float32)

# Encode the target variable
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

print(f"\nTarget classes mapping:")
for i, class_name in enumerate(label_encoder.classes_):
    print(f"  {i}: {class_name}")

# Split the data
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

print(f"\nTraining set: {X_train.shape}")
print(f"Test set: {X_test.shape}")

# Train Random Forest model
print("\nTraining Random Forest model...")
rf_model = RandomForestClassifier(
    n_estimators=100,
    random_state=42,
    max_depth=15,
    min_samples_split=5,
    min_samples_leaf=2,
    n_jobs=-1
)

rf_model.fit(X_train, y_train)
print("Model training completed successfully!")

# Make predictions
y_pred = rf_model.predict(X_test)
y_pred_proba = rf_model.predict_proba(X_test)

# ==================== COMPREHENSIVE METRICS CALCULATION ====================

print("\n" + "="*60)
print("COMPREHENSIVE MODEL EVALUATION METRICS")
print("="*60)

# Basic accuracy
accuracy = accuracy_score(y_test, y_pred)

# Precision, Recall, F1 for each class
precision_per_class = precision_score(y_test, y_pred, average=None, zero_division=0)
recall_per_class = recall_score(y_test, y_pred, average=None, zero_division=0)
f1_per_class = f1_score(y_test, y_pred, average=None, zero_division=0)

# Weighted and macro averages
precision_weighted = precision_score(y_test, y_pred, average='weighted', zero_division=0)
recall_weighted = recall_score(y_test, y_pred, average='weighted', zero_division=0)
f1_weighted = f1_score(y_test, y_pred, average='weighted', zero_division=0)

precision_macro = precision_score(y_test, y_pred, average='macro', zero_division=0)
recall_macro = recall_score(y_test, y_pred, average='macro', zero_division=0)
f1_macro = f1_score(y_test, y_pred, average='macro', zero_division=0)

# Calculate False Positive Rate (FPR) for each class
cm = confusion_matrix(y_test, y_pred)
fpr_per_class = []
for i in range(len(label_encoder.classes_)):
    fp = cm[:, i].sum() - cm[i, i]
    tn = cm.sum() - (cm[i, :].sum() + cm[:, i].sum() - cm[i, i])
    fpr = fp / (fp + tn) if (fp + tn) > 0 else 0
    fpr_per_class.append(fpr)

# Create comprehensive metrics dataframe
metrics_df = pd.DataFrame({
    'Class': label_encoder.classes_,
    'Precision': precision_per_class,
    'Recall': recall_per_class,
    'F1-Score': f1_per_class,
    'False Positive Rate': fpr_per_class
})

# Add overall metrics
overall_metrics = pd.DataFrame({
    'Class': ['Weighted Avg', 'Macro Avg'],
    'Precision': [precision_weighted, precision_macro],
    'Recall': [recall_weighted, recall_macro],
    'F1-Score': [f1_weighted, f1_macro],
    'False Positive Rate': [np.nan, np.nan]  # FPR not typically calculated for averages
})

metrics_df = pd.concat([metrics_df, overall_metrics], ignore_index=True)

print(f"\nOverall Accuracy: {accuracy:.4f}")
print(f"\nDetailed Metrics per Class:")
print(metrics_df.round(4))

print(f"\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=label_encoder.classes_))

# ==================== VISUALIZATIONS ====================

print("\nGenerating visualizations...")

# 1. Confusion Matrix Heatmap
plt.figure(figsize=(12, 10))
cm = confusion_matrix(y_test, y_pred)
cm_percentage = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

sns.heatmap(cm_percentage, annot=True, fmt='.2%', cmap='Blues', 
            xticklabels=label_encoder.classes_, 
            yticklabels=label_encoder.classes_)
plt.title('Confusion Matrix Heatmap (Percentage)', fontsize=16, fontweight='bold')
plt.xlabel('Predicted Label', fontsize=12)
plt.ylabel('True Label', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.savefig('confusion_matrix_heatmap.png', dpi=300, bbox_inches='tight')
plt.show()

# 2. Metrics Comparison Heatmap (FIXED THE ERROR)
plt.figure(figsize=(12, 8))
metrics_heatmap_data = metrics_df.iloc[:-2][['Precision', 'Recall', 'F1-Score', 'False Positive Rate']]
metrics_heatmap_data.index = metrics_df.iloc[:-2]['Class']

sns.heatmap(metrics_heatmap_data.T, annot=True, fmt='.3f', cmap='viridis', 
            center=0.5, cbar_kws={'label': 'Score'})  # Fixed typo here
plt.title('Performance Metrics Heatmap by Class', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.savefig('metrics_heatmap.png', dpi=300, bbox_inches='tight')
plt.show()

# 3. Pie Chart for Class Distribution in Predictions
plt.figure(figsize=(15, 6))

# Actual distribution
plt.subplot(1, 2, 1)
actual_counts = pd.Series(y_test).value_counts()
actual_labels = [label_encoder.classes_[i] for i in actual_counts.index]
plt.pie(actual_counts.values, labels=actual_labels, autopct='%1.1f%%', startangle=90)
plt.title('Actual Class Distribution (Test Set)', fontsize=14, fontweight='bold')

# Predicted distribution
plt.subplot(1, 2, 2)
predicted_counts = pd.Series(y_pred).value_counts()
predicted_labels = [label_encoder.classes_[i] for i in predicted_counts.index]
plt.pie(predicted_counts.values, labels=predicted_labels, autopct='%1.1f%%', startangle=90)
plt.title('Predicted Class Distribution', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.savefig('class_distribution_piechart.png', dpi=300, bbox_inches='tight')
plt.show()

# 4. Additional Pie Chart: Training Data Distribution
plt.figure(figsize=(15, 6))

# Training data distribution
plt.subplot(1, 2, 1)
train_counts = pd.Series(y_train).value_counts()
train_labels = [label_encoder.classes_[i] for i in train_counts.index]
colors = plt.cm.Set3(np.linspace(0, 1, len(train_labels)))
plt.pie(train_counts.values, labels=train_labels, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title('Training Data Class Distribution', fontsize=14, fontweight='bold')

# Overall dataset distribution
plt.subplot(1, 2, 2)
overall_counts = df[' Label'].value_counts()
plt.pie(overall_counts.values, labels=overall_counts.index, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title('Overall Dataset Class Distribution', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.savefig('data_distribution_piechart.png', dpi=300, bbox_inches='tight')
plt.show()

# 5. TP, TN, FP, FN Pie Chart (NEW ADDITION)
plt.figure(figsize=(10, 8))

# Calculate TP, TN, FP, FN for binary classification (BENIGN vs all attacks)
# Treat BENIGN as positive class (index 0) and all attacks as negative class
benign_class_index = np.where(label_encoder.classes_ == 'BENIGN')[0][0]

# Create binary labels (1 for BENIGN, 0 for attacks)
y_test_binary = (y_test == benign_class_index).astype(int)
y_pred_binary = (y_pred == benign_class_index).astype(int)

# Calculate TP, TN, FP, FN for binary classification
from sklearn.metrics import confusion_matrix
tn, fp, fn, tp = confusion_matrix(y_test_binary, y_pred_binary).ravel()

# Create pie chart for TP, TN, FP, FN
categories = ['True Positives', 'True Negatives', 'False Positives', 'False Negatives']
values = [tp, tn, fp, fn]
colors = ['#2ecc71', '#3498db', '#e74c3c', '#f39c12']  # Green, Blue, Red, Orange

plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=90, colors=colors, 
        explode=(0.05, 0.05, 0.05, 0.05), shadow=True)
plt.title('TP, TN, FP, FN Distribution\n(BENIGN vs All Attacks)', fontsize=16, fontweight='bold')

# Add text box with counts
textstr = f'TP: {tp}\nTN: {tn}\nFP: {fp}\nFN: {fn}\nTotal: {tp+tn+fp+fn}'
props = dict(boxstyle='round', facecolor='wheat', alpha=0.8)
plt.gcf().text(0.85, 0.15, textstr, fontsize=12, bbox=props)

plt.tight_layout()
plt.savefig('tp_tn_fp_fn_piechart.png', dpi=300, bbox_inches='tight')
plt.show()

print(f"\nBinary Classification Metrics (BENIGN vs All Attacks):")
print(f"True Positives (TP): {tp}")
print(f"True Negatives (TN): {tn}")
print(f"False Positives (FP): {fp}")
print(f"False Negatives (FN): {fn}")
print(f"Accuracy: {(tp + tn) / (tp + tn + fp + fn):.4f}")

# 6. Feature Importance Plot
plt.figure(figsize=(12, 8))
feature_importance = pd.DataFrame({
    'feature': feature_columns,
    'importance': rf_model.feature_importances_
}).sort_values('importance', ascending=False)

top_20_features = feature_importance.head(20)
sns.barplot(data=top_20_features, x='importance', y='feature')
plt.title('Top 20 Most Important Features', fontsize=16, fontweight='bold')
plt.xlabel('Feature Importance Score', fontsize=12)
plt.tight_layout()
plt.savefig('feature_importance.png', dpi=300, bbox_inches='tight')
plt.show()

# 7. Metrics Bar Plot by Class
plt.figure(figsize=(14, 8))
metrics_by_class = metrics_df.iloc[:-2].set_index('Class')
metrics_by_class[['Precision', 'Recall', 'F1-Score']].plot(kind='bar', figsize=(14, 8))
plt.title('Precision, Recall, and F1-Score by Class', fontsize=16, fontweight='bold')
plt.xlabel('Class', fontsize=12)
plt.ylabel('Score', fontsize=12)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('metrics_by_class.png', dpi=300, bbox_inches='tight')
plt.show()

# 8. False Positive Rate Pie Chart
plt.figure(figsize=(10, 8))
fpr_data = metrics_df.iloc[:-2].set_index('Class')['False Positive Rate']
colors = plt.cm.Reds(np.linspace(0.3, 0.8, len(fpr_data)))

# Create a pie chart showing FPR distribution
plt.pie(fpr_data.values, labels=fpr_data.index, autopct='%1.2f%%', 
        startangle=90, colors=colors, shadow=True)
plt.title('False Positive Rate Distribution by Class', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.savefig('fpr_piechart.png', dpi=300, bbox_inches='tight')
plt.show()

# ==================== SAVE MODEL AND METADATA ====================

print("\nSaving model and metadata...")

# Save the trained model
model_filename = 'random_forest_ddos_model.joblib'
joblib.dump(rf_model, model_filename)
print(f"Model saved as: {model_filename}")

# Save the label encoder
encoder_filename = 'label_encoder.joblib'
joblib.dump(label_encoder, encoder_filename)
print(f"Label encoder saved as: {encoder_filename}")

# Save feature names
feature_names_filename = 'feature_names.joblib'
joblib.dump(feature_columns, feature_names_filename)
print(f"Feature names saved as: {feature_names_filename}")

# Save comprehensive metrics to CSV
metrics_df.to_csv('model_metrics_detailed.csv', index=False)
print("Detailed metrics saved as: model_metrics_detailed.csv")

# Save summary report
with open('model_summary_report.txt', 'w') as f:
    f.write("RANDOM FOREST DDoS DETECTION MODEL - SUMMARY REPORT\n")
    f.write("=" * 50 + "\n\n")
    f.write(f"Model: Random Forest Classifier\n")
    f.write(f"Training Date: {pd.Timestamp.now()}\n")
    f.write(f"Dataset Shape: {df.shape}\n")
    f.write(f"Number of Features: {len(feature_columns)}\n")
    f.write(f"Classes: {list(label_encoder.classes_)}\n\n")
    
    f.write("PERFORMANCE METRICS\n")
    f.write("-" * 30 + "\n")
    f.write(f"Overall Accuracy: {accuracy:.4f}\n")
    f.write(f"Weighted Precision: {precision_weighted:.4f}\n")
    f.write(f"Weighted Recall: {recall_weighted:.4f}\n")
    f.write(f"Weighted F1-Score: {f1_weighted:.4f}\n")
    f.write(f"Macro Precision: {precision_macro:.4f}\n")
    f.write(f"Macro Recall: {recall_macro:.4f}\n")
    f.write(f"Macro F1-Score: {f1_macro:.4f}\n\n")
    
    f.write("BINARY CLASSIFICATION (BENIGN vs ATTACKS)\n")
    f.write("-" * 40 + "\n")
    f.write(f"True Positives (TP): {tp}\n")
    f.write(f"True Negatives (TN): {tn}\n")
    f.write(f"False Positives (FP): {fp}\n")
    f.write(f"False Negatives (FN): {fn}\n")
    f.write(f"Binary Accuracy: {(tp + tn) / (tp + tn + fp + fn):.4f}\n\n")
    
    f.write("TOP 10 FEATURES\n")
    f.write("-" * 30 + "\n")
    for i, row in feature_importance.head(10).iterrows():
        f.write(f"{i+1:2d}. {row['feature']:30} {row['importance']:.4f}\n")

print("Summary report saved as: model_summary_report.txt")

print("\n" + "="*60)
print("ALL FILES SAVED SUCCESSFULLY!")
print("="*60)
print("Model files created:")
print(f"  - {model_filename} (trained model)")
print(f"  - {encoder_filename} (label encoder)")
print(f"  - {feature_names_filename} (feature names)")
print("\nVisualization files created:")
print("  - confusion_matrix_heatmap.png")
print("  - metrics_heatmap.png")
print("  - class_distribution_piechart.png")
print("  - data_distribution_piechart.png")
print("  - tp_tn_fp_fn_piechart.png")  # NEW PIE CHART
print("  - feature_importance.png")
print("  - metrics_by_class.png")
print("  - fpr_piechart.png")
print("\nData files created:")
print("  - model_metrics_detailed.csv")
print("  - model_summary_report.txt")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import (classification_report, confusion_matrix, accuracy_score, 
                           precision_score, recall_score, f1_score, precision_recall_fscore_support)
import warnings
warnings.filterwarnings('ignore')

# Set style for better plots
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

# Load your balanced dataset
df = pd.read_csv('balanced_dataset.csv')

# Display dataset info
print("Dataset shape:", df.shape)
print("\nLabel distribution:")
print(df[' Label'].value_counts())

# Identify columns to exclude
exclude_columns = [
    'Unnamed: 0',      # Index column
    'Flow ID',         # Flow identifier
    ' Source IP',      # IP address
    ' Destination IP', # IP address
    ' Timestamp',      # Timestamp
    ' Label'           # This is our target variable
]

# Features to use (all remaining columns except the excluded ones)
feature_columns = [col for col in df.columns if col not in exclude_columns]

print(f"\nUsing {len(feature_columns)} features:")

# Prepare features and target
X = df[feature_columns]
y = df[' Label']

# Data Cleaning Steps
print("\n=== DATA CLEANING ===")

# 1. Convert non-numeric columns to numeric
non_numeric_cols = X.select_dtypes(include=['object']).columns.tolist()
if non_numeric_cols:
    print(f"Converting non-numeric columns: {non_numeric_cols}")
    for col in non_numeric_cols:
        X[col] = pd.to_numeric(X[col], errors='coerce')

# 2. Check for infinite values and replace them
print("Checking for infinite values...")
inf_mask = np.isinf(X.select_dtypes(include=[np.number]))
if inf_mask.any().any():
    inf_columns = inf_mask.any()
    inf_cols_list = inf_columns[inf_columns].index.tolist()
    print(f"Found infinite values in columns: {inf_cols_list}")
    
    # Replace inf with NaN
    X = X.replace([np.inf, -np.inf], np.nan)

# 3. Handle extreme values by clipping
print("Handling extreme values...")
for col in X.select_dtypes(include=[np.number]).columns:
    Q1 = X[col].quantile(0.25)
    Q3 = X[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 3 * IQR
    upper_bound = Q3 + 3 * IQR
    
    # Clip extreme outliers
    if not np.isnan(lower_bound) and not np.isnan(upper_bound):
        outliers_before = ((X[col] < lower_bound) | (X[col] > upper_bound)).sum()
        if outliers_before > 0:
            print(f"  {col}: clipping {outliers_before} outliers")
            X[col] = np.clip(X[col], lower_bound, upper_bound)

# 4. Fill NaN values with median
print("Filling NaN values...")
nan_before = X.isnull().sum().sum()
print(f"NaN values before filling: {nan_before}")

for col in X.columns:
    if X[col].isnull().sum() > 0:
        X[col].fillna(X[col].median(), inplace=True)

print(f"NaN values after filling: {X.isnull().sum().sum()}")

# 5. Scale large values to prevent float32 overflow
print("Scaling large values...")
for col in X.select_dtypes(include=[np.number]).columns:
    col_max = abs(X[col]).max()
    if col_max > 1e6:
        print(f"  Scaling {col} (max absolute value: {col_max:.2e})")
        X[col] = X[col] / col_max

# Convert to float32
X = X.astype(np.float32)

# Encode the target variable
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

print(f"\nTarget classes mapping:")
for i, class_name in enumerate(label_encoder.classes_):
    print(f"  {i}: {class_name}")

# Split the data
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

print(f"\nTraining set: {X_train.shape}")
print(f"Test set: {X_test.shape}")

# Train Random Forest model
print("\nTraining Random Forest model...")
rf_model = RandomForestClassifier(
    n_estimators=100,
    random_state=42,
    max_depth=15,
    min_samples_split=5,
    min_samples_leaf=2,
    n_jobs=-1
)

rf_model.fit(X_train, y_train)
print("Model training completed successfully!")

# Make predictions
y_pred = rf_model.predict(X_test)
y_pred_proba = rf_model.predict_proba(X_test)

# ==================== COMPREHENSIVE METRICS CALCULATION ====================

print("\n" + "="*60)
print("COMPREHENSIVE MODEL EVALUATION METRICS")
print("="*60)

# Basic accuracy
accuracy = accuracy_score(y_test, y_pred)

# Precision, Recall, F1 for each class
precision_per_class = precision_score(y_test, y_pred, average=None, zero_division=0)
recall_per_class = recall_score(y_test, y_pred, average=None, zero_division=0)
f1_per_class = f1_score(y_test, y_pred, average=None, zero_division=0)

# Weighted and macro averages
precision_weighted = precision_score(y_test, y_pred, average='weighted', zero_division=0)
recall_weighted = recall_score(y_test, y_pred, average='weighted', zero_division=0)
f1_weighted = f1_score(y_test, y_pred, average='weighted', zero_division=0)

precision_macro = precision_score(y_test, y_pred, average='macro', zero_division=0)
recall_macro = recall_score(y_test, y_pred, average='macro', zero_division=0)
f1_macro = f1_score(y_test, y_pred, average='macro', zero_division=0)

# Calculate False Positive Rate (FPR) for each class
cm = confusion_matrix(y_test, y_pred)
fpr_per_class = []
for i in range(len(label_encoder.classes_)):
    fp = cm[:, i].sum() - cm[i, i]
    tn = cm.sum() - (cm[i, :].sum() + cm[:, i].sum() - cm[i, i])
    fpr = fp / (fp + tn) if (fp + tn) > 0 else 0
    fpr_per_class.append(fpr)

# Create comprehensive metrics dataframe
metrics_df = pd.DataFrame({
    'Class': label_encoder.classes_,
    'Precision': precision_per_class,
    'Recall': recall_per_class,
    'F1-Score': f1_per_class,
    'False Positive Rate': fpr_per_class
})

# Add overall metrics
overall_metrics = pd.DataFrame({
    'Class': ['Weighted Avg', 'Macro Avg'],
    'Precision': [precision_weighted, precision_macro],
    'Recall': [recall_weighted, recall_macro],
    'F1-Score': [f1_weighted, f1_macro],
    'False Positive Rate': [np.nan, np.nan]  # FPR not typically calculated for averages
})

metrics_df = pd.concat([metrics_df, overall_metrics], ignore_index=True)

print(f"\nOverall Accuracy: {accuracy:.4f}")
print(f"\nDetailed Metrics per Class:")
print(metrics_df.round(4))

print(f"\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=label_encoder.classes_))

# ==================== TP, TN, FP, FN TEXTUAL PERCENTAGES ====================

print("\n" + "="*60)
print("BINARY CLASSIFICATION METRICS (BENIGN vs ALL ATTACKS)")
print("="*60)

# Calculate TP, TN, FP, FN for binary classification (BENIGN vs all attacks)
benign_class_index = np.where(label_encoder.classes_ == 'BENIGN')[0][0]

# Create binary labels (1 for BENIGN, 0 for attacks)
y_test_binary = (y_test == benign_class_index).astype(int)
y_pred_binary = (y_pred == benign_class_index).astype(int)

# Calculate TP, TN, FP, FN for binary classification
from sklearn.metrics import confusion_matrix
tn, fp, fn, tp = confusion_matrix(y_test_binary, y_pred_binary).ravel()

total_samples = tp + tn + fp + fn

# Calculate percentages
tp_percentage = (tp / total_samples) * 100
tn_percentage = (tn / total_samples) * 100
fp_percentage = (fp / total_samples) * 100
fn_percentage = (fn / total_samples) * 100

# Display textual percentages
print(f"\nDIRECT PERCENTAGES OF TP, TN, FP, FN:")
print(f"True Positives (TP)  : {tp:4d} samples ({tp_percentage:6.2f}%) - Correctly identified BENIGN traffic")
print(f"True Negatives (TN)  : {tn:4d} samples ({tn_percentage:6.2f}%) - Correctly identified ATTACK traffic")
print(f"False Positives (FP) : {fp:4d} samples ({fp_percentage:6.2f}%) - Attacks misclassified as BENIGN")
print(f"False Negatives (FN) : {fn:4d} samples ({fn_percentage:6.2f}%) - BENIGN misclassified as attacks")
print(f"Total Samples        : {total_samples:4d} samples (100.00%)")

# Calculate additional binary metrics
binary_accuracy = (tp + tn) / total_samples * 100
misclassification_rate = (fp + fn) / total_samples * 100

print(f"\nBINARY CLASSIFICATION SUMMARY:")
print(f"Correct Predictions  : {tp + tn:4d} samples ({binary_accuracy:6.2f}%)")
print(f"Wrong Predictions    : {fp + fn:4d} samples ({misclassification_rate:6.2f}%)")
print(f"Binary Accuracy      : {binary_accuracy:.2f}%")

# Calculate precision and recall for binary classification
binary_precision = tp / (tp + fp) * 100 if (tp + fp) > 0 else 0
binary_recall = tp / (tp + fn) * 100 if (tp + fn) > 0 else 0
binary_f1 = 2 * (binary_precision * binary_recall) / (binary_precision + binary_recall) if (binary_precision + binary_recall) > 0 else 0

print(f"\nBINARY CLASSIFICATION METRICS:")
print(f"Precision (BENIGN)   : {binary_precision:6.2f}% - How many predicted BENIGN are actually BENIGN")
print(f"Recall (BENIGN)      : {binary_recall:6.2f}% - How many actual BENIGN are correctly identified")
print(f"F1-Score (BENIGN)    : {binary_f1:6.2f}%")

# ==================== VISUALIZATIONS ====================

print("\nGenerating visualizations...")

# 1. Confusion Matrix Heatmap
plt.figure(figsize=(12, 10))
cm = confusion_matrix(y_test, y_pred)
cm_percentage = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

sns.heatmap(cm_percentage, annot=True, fmt='.2%', cmap='Blues', 
            xticklabels=label_encoder.classes_, 
            yticklabels=label_encoder.classes_)
plt.title('Confusion Matrix Heatmap (Percentage)', fontsize=16, fontweight='bold')
plt.xlabel('Predicted Label', fontsize=12)
plt.ylabel('True Label', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.savefig('confusion_matrix_heatmap.png', dpi=300, bbox_inches='tight')
plt.show()

# 2. Metrics Comparison Heatmap
plt.figure(figsize=(12, 8))
metrics_heatmap_data = metrics_df.iloc[:-2][['Precision', 'Recall', 'F1-Score', 'False Positive Rate']]
metrics_heatmap_data.index = metrics_df.iloc[:-2]['Class']

sns.heatmap(metrics_heatmap_data.T, annot=True, fmt='.3f', cmap='viridis', 
            center=0.5, cbar_kws={'label': 'Score'})
plt.title('Performance Metrics Heatmap by Class', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.savefig('metrics_heatmap.png', dpi=300, bbox_inches='tight')
plt.show()

# 3. Pie Chart for Class Distribution in Predictions
plt.figure(figsize=(15, 6))

# Actual distribution
plt.subplot(1, 2, 1)
actual_counts = pd.Series(y_test).value_counts()
actual_labels = [label_encoder.classes_[i] for i in actual_counts.index]
plt.pie(actual_counts.values, labels=actual_labels, autopct='%1.1f%%', startangle=90)
plt.title('Actual Class Distribution (Test Set)', fontsize=14, fontweight='bold')

# Predicted distribution
plt.subplot(1, 2, 2)
predicted_counts = pd.Series(y_pred).value_counts()
predicted_labels = [label_encoder.classes_[i] for i in predicted_counts.index]
plt.pie(predicted_counts.values, labels=predicted_labels, autopct='%1.1f%%', startangle=90)
plt.title('Predicted Class Distribution', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.savefig('class_distribution_piechart.png', dpi=300, bbox_inches='tight')
plt.show()

# 4. TP, TN, FP, FN Pie Chart
plt.figure(figsize=(12, 8))

# Create pie chart for TP, TN, FP, FN
categories = ['True Positives', 'True Negatives', 'False Positives', 'False Negatives']
values = [tp, tn, fp, fn]
colors = ['#2ecc71', '#3498db', '#e74c3c', '#f39c12']  # Green, Blue, Red, Orange
percentages = [tp_percentage, tn_percentage, fp_percentage, fn_percentage]

# Create labels with both count and percentage
labels = [f'{cat}\n{val} samples ({pct:.1f}%)' 
          for cat, val, pct in zip(categories, values, percentages)]

plt.pie(values, labels=labels, startangle=90, colors=colors, 
        explode=(0.05, 0.05, 0.05, 0.05), shadow=True)
plt.title('TP, TN, FP, FN Distribution\n(BENIGN vs All Attacks)', fontsize=16, fontweight='bold')

plt.tight_layout()
plt.savefig('tp_tn_fp_fn_piechart.png', dpi=300, bbox_inches='tight')
plt.show()

# 5. Feature Importance Plot
plt.figure(figsize=(12, 8))
feature_importance = pd.DataFrame({
    'feature': feature_columns,
    'importance': rf_model.feature_importances_
}).sort_values('importance', ascending=False)

top_20_features = feature_importance.head(20)
sns.barplot(data=top_20_features, x='importance', y='feature')
plt.title('Top 20 Most Important Features', fontsize=16, fontweight='bold')
plt.xlabel('Feature Importance Score', fontsize=12)
plt.tight_layout()
plt.savefig('feature_importance.png', dpi=300, bbox_inches='tight')
plt.show()

# 6. Metrics Bar Plot by Class
plt.figure(figsize=(14, 8))
metrics_by_class = metrics_df.iloc[:-2].set_index('Class')
metrics_by_class[['Precision', 'Recall', 'F1-Score']].plot(kind='bar', figsize=(14, 8))
plt.title('Precision, Recall, and F1-Score by Class', fontsize=16, fontweight='bold')
plt.xlabel('Class', fontsize=12)
plt.ylabel('Score', fontsize=12)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('metrics_by_class.png', dpi=300, bbox_inches='tight')
plt.show()

# ==================== SAVE MODEL AND METADATA ====================

print("\nSaving model and metadata...")

# Save the trained model
model_filename = 'random_forest_ddos_model.joblib'
joblib.dump(rf_model, model_filename)
print(f"Model saved as: {model_filename}")

# Save the label encoder
encoder_filename = 'label_encoder.joblib'
joblib.dump(label_encoder, encoder_filename)
print(f"Label encoder saved as: {encoder_filename}")

# Save feature names
feature_names_filename = 'feature_names.joblib'
joblib.dump(feature_columns, feature_names_filename)
print(f"Feature names saved as: {feature_names_filename}")

# Save comprehensive metrics to CSV
metrics_df.to_csv('model_metrics_detailed.csv', index=False)
print("Detailed metrics saved as: model_metrics_detailed.csv")

# Save binary classification metrics
binary_metrics = pd.DataFrame({
    'Metric': ['True Positives', 'True Negatives', 'False Positives', 'False Negatives', 
               'Total Samples', 'Binary Accuracy', 'Precision', 'Recall', 'F1-Score'],
    'Count': [tp, tn, fp, fn, total_samples, tp + tn, tp, tp, np.nan],
    'Percentage': [tp_percentage, tn_percentage, fp_percentage, fn_percentage, 
                   100, binary_accuracy, binary_precision, binary_recall, binary_f1]
})
binary_metrics.to_csv('binary_classification_metrics.csv', index=False)
print("Binary metrics saved as: binary_classification_metrics.csv")

# Save summary report
with open('model_summary_report.txt', 'w') as f:
    f.write("RANDOM FOREST DDoS DETECTION MODEL - SUMMARY REPORT\n")
    f.write("=" * 50 + "\n\n")
    f.write(f"Model: Random Forest Classifier\n")
    f.write(f"Training Date: {pd.Timestamp.now()}\n")
    f.write(f"Dataset Shape: {df.shape}\n")
    f.write(f"Number of Features: {len(feature_columns)}\n")
    f.write(f"Classes: {list(label_encoder.classes_)}\n\n")
    
    f.write("PERFORMANCE METRICS\n")
    f.write("-" * 30 + "\n")
    f.write(f"Overall Accuracy: {accuracy:.4f}\n")
    f.write(f"Weighted Precision: {precision_weighted:.4f}\n")
    f.write(f"Weighted Recall: {recall_weighted:.4f}\n")
    f.write(f"Weighted F1-Score: {f1_weighted:.4f}\n")
    f.write(f"Macro Precision: {precision_macro:.4f}\n")
    f.write(f"Macro Recall: {recall_macro:.4f}\n")
    f.write(f"Macro F1-Score: {f1_macro:.4f}\n\n")
    
    f.write("BINARY CLASSIFICATION (BENIGN vs ATTACKS)\n")
    f.write("-" * 40 + "\n")
    f.write(f"True Positives (TP): {tp} samples ({tp_percentage:.2f}%)\n")
    f.write(f"True Negatives (TN): {tn} samples ({tn_percentage:.2f}%)\n")
    f.write(f"False Positives (FP): {fp} samples ({fp_percentage:.2f}%)\n")
    f.write(f"False Negatives (FN): {fn} samples ({fn_percentage:.2f}%)\n")
    f.write(f"Binary Accuracy: {binary_accuracy:.2f}%\n")
    f.write(f"Precision (BENIGN): {binary_precision:.2f}%\n")
    f.write(f"Recall (BENIGN): {binary_recall:.2f}%\n")
    f.write(f"F1-Score (BENIGN): {binary_f1:.2f}%\n\n")
    
    f.write("TOP 10 FEATURES\n")
    f.write("-" * 30 + "\n")
    for i, row in feature_importance.head(10).iterrows():
        f.write(f"{i+1:2d}. {row['feature']:30} {row['importance']:.4f}\n")

print("Summary report saved as: model_summary_report.txt")

print("\n" + "="*60)
print("ALL FILES SAVED SUCCESSFULLY!")
print("="*60)
print("Model files created:")
print(f"  - {model_filename} (trained model)")
print(f"  - {encoder_filename} (label encoder)")
print(f"  - {feature_names_filename} (feature names)")
print("\nVisualization files created:")
print("  - confusion_matrix_heatmap.png")
print("  - metrics_heatmap.png")
print("  - class_distribution_piechart.png")
print("  - tp_tn_fp_fn_piechart.png")
print("  - feature_importance.png")
print("  - metrics_by_class.png")
print("\nData files created:")
print("  - model_metrics_detailed.csv")
print("  - binary_classification_metrics.csv")
print("  - model_summary_report.txt")
```
